### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"
```

```python
datos = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda1["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

print("Ingreso total de la Tienda 1:", ingreso_tienda_1)
print("Ingreso total de la Tienda 2:", ingreso_tienda_2)
print("Ingreso total de la Tienda 3:", ingreso_tienda_3)
print("Ingreso total de la Tienda 4:", ingreso_tienda_4)
```

```python
datos.groupby("Tienda")["Precio"].sum()
```

# 3. Calificación promedio de la tienda

## Gráfico de dispersión de ventas por tienda

```python
plt.figure(figsize=(10, 7))

for tienda, df in datos.groupby("Tienda"):
    plt.scatter(df["lon"], df["lat"], label=tienda, alpha=0.5)

plt.title("Distribución geográfica de las ventas por tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.show()
```

## Mapa de Calor

```python
plt.figure(figsize=(10, 7))

plt.hist2d(datos["lon"], datos["lat"], bins=40)

plt.colorbar(label="Densidad de ventas")
plt.title("Mapa de calor de concentración de ventas")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.show()
```
